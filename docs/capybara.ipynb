## Grayscale Function Vignette

This vignette demonstrates the use of the `grayscale(image)` function in the ImaGenie package. We will use the image of a capybara (`capybara.jpeg`) for our examples.


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from imagenie.grayscale import grayscale

# Load the capybara image
image_path = "img/capybara.jpeg"
image = np.array(Image.open(image_path))

# Display the original image
plt.imshow(image)
plt.title("Original Capybara Image")
plt.axis("off")
plt.show()
```

```python
# Convert the image to grayscale
gray_image = grayscale(image)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Capybara Image")
plt.axis("off")
plt.show()
```
