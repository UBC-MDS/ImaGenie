# Image tutorial

A short tutorial for on plotting images using ImaGenie.

In this tutorial, youâ€™ll learn a 4 techniques for creating image transformations in ImaGenie. The four techniques are grayscale, flip, blur and scale.

For this tutorial, we will create visualizations to explore transformed images of cute capybaras.

ImaGenie is designed to work with data in the form of numpy arrays. Let's import the four cute capybara images and convert them into numpy arrays accordingly. 

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import imagenie
from imagenie.augment import augment
from imagenie.blur import blur
from imagenie.flip import flip
from imagenie.grayscale import grayscale
from imagenie.scale import scale
# Load the capybara image and convert them to numpy arrays
image_1 = plt.imread("img/capybara.jpeg")
image_2 = plt.imread("img/capybara2.jpeg")
image_3 = plt.imread("img/capybara3.jpeg")
image_4 = plt.imread("img/capybara4.jpeg")
img=plt.imread("../tests/testimage.png")
images=[image_1,image_2,image_3,image_4]
```

```python
#Define pixel-DPI ratio to calculate display size
px = 1/plt.rcParams['figure.dpi'] 
```

```python
titles = ['Appybara', 'Gangsta Capybara', 'Capybarista', 'Luckybara']
for i,j in enumerate(images,start=1):
    plt.subplot(2,2,i)
    plt.imshow(j)
    plt.title(titles[i-1])
    plt.axis("off")
    plt.plot()
plt.tight_layout()
```

# Blur Function

Adds a Gaussian Blur by adding noise to the input image leveraging the Gaussian distribution with a provided value of standard deviation


```python
blurred_img=blur(np.asarray(image_3),2)
plt.imshow(blurred_img)
plt.title("Blurred Capybarista")
plt.axis("off")
plt.show()
```

# Scale Function

The scale function scales an image's numpy array using a scaling factor N. The scaling factor N is a positive float number. 

- Make the image 1.5 times larger:

```python
large_image = scale(image_4, 1.5)

# Display the larger image
plt.figure(figsize=(large_image.shape[1]*px,large_image.shape[0]*px)) # Added code to display resized image in actual size
plt.imshow(large_image)
plt.title("Bigger Luckybara")
plt.axis("off")
plt.show()
```

- Make the image 0.5 times smaller:

```python
small_image = scale(image_4, 0.5)

# Display the smaller image
plt.figure(figsize=(small_image.shape[1]*px,small_image.shape[0]*px)) # Added code to display resized image in actual size
plt.imshow(small_image)
plt.title("Smaller Luckybara")
plt.axis("off")
plt.show()
```

## Grayscale Function Vignette

This vignette demonstrates the use of the `grayscale(image)` function in the ImaGenie package. We will use the image of a capybara (`capybara.jpeg`) for our examples.


```python
# Convert the image to grayscale
gray_image = grayscale(image_1)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscaled Appybara (he's sad)")
plt.axis("off")
plt.show()
```
